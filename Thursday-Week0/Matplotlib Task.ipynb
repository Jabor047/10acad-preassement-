```python
plt.boxplot(data['Q1'], notch=True)
plt.title("A Boxplot showing the Age from the dataset")
plt.show();
```

```python
plt.bar(data['Q2'], data['Q3'], align='edge')
plt.title("A Bar Plot comparing the Marriage status between the two sexes")
plt.xlabel("Sex")
plt.ylabel("Marital Status")
plt.show()
```

```python
plt.hist(data['Q1'], bins=50, histtype='barstacked', color='black')
plt.title("A Histogram Showing The Age Distribution In The Data")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show;
```
